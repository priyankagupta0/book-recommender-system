## Popularity based recommender system.

```python
ratings_books=ratings.merge(books, on ='ISBN')
```

```python
ratings_books.shape
```

```python
#we have one book with many different rating. we will group them
```

```python
num_ratings_df= ratings_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_ratings_df
```

```python
#average ratings of each book
```

```python
avg_ratings_df= ratings_books.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_ratings_df
```

```python
popular_df= avg_ratings_df.merge(num_ratings_df, on='Book-Title')
popular_df
```

```python
popular_df =popular_df[popular_df['num_ratings']>300].sort_values('avg_ratings',ascending=False).head(100)
popular_df
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','avg_ratings','num_ratings','Book-Author','Year-Of-Publication','Publisher','Image-URL-M']]
popular_df
```

## collaborative filtering based recommendation system

```python
x=ratings_books.groupby('User-ID').count()['Book-Rating']>200
good_users=x[x].index
```

```python
filtered_ratings= ratings_books[ratings_books['User-ID'].isin(good_users)]
filtered_ratings
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_book_titles= y[y].index
famous_book_titles
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_book_titles)]
```

```python
pt= final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
#check the simialrity of each book with other books
from sklearn.metrics.pairwise import cosine_similarity
similarity_score= cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(Book_Title):
    #find the place of the index. [0]th item ka [0]th 
    index = np.where(pt.index== Book_Title)[0][0]
    
    #enumerates gives us the value along with its index
    similar_items= sorted(list(enumerate(similarity_score[index])),key= lambda x:x[1], reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
#similarity_score[0]# similarity of 0 index attribute with every index.
```
